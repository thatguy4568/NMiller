Trip Distances

To calculate the distance of each trip, we use the latitude and longitude of the starting and ending stations. We apply the Haversine Formula which determines the shortest distance between two points on earth also known as " the crow flies" distance. Although the distance is accurate when stations are on the same street or avenue, it may not be as precise when they are not. However, it provides a minimum distance estimation between two stations. This method is used because Citibikes do not have built-in GPS systems, making it impossible to capture the exact route length of a rider.

```python
def haversine(lat1, lon1, lat2, lon2):
    lon1, lat1, lon2, lat2 = \
    map(np.radians ,[lon1, lat1, lon2, lat2])
    h = np.sin((lat2-lat1)/2.0)**2 + np.cos(lat1) * np.cos(lat2) * np.sin((lon2-lon1)/2.0)**2
    miles = 3959 * (2 * np.arcsin(np.sqrt(h)))
    return miles
```

Below is two graphs that shows the count of trips calculated by distance in miles. The chart shows us that there a lot of trips with a distance of zero! It turns out that even though Citi Bike is intended for transportation it also gets used for recreation where trips start and end at the same station.

```python
#Define the .csv file
bike_df = pd.read_csv('/content/gdrive/MyDrive/final_project_data/202012-citibike-tripdata.csv')

# Calculate haversine distances
distance = haversine(bike_df['start station latitude'], bike_df['start station longitude'],
                     bike_df['end station latitude'], bike_df['end station longitude'])
bike_df['distance'] = distance

# Visualize the distribution of distances without filtering
sns.displot(data=bike_df, x="distance", aspect=10/5)
plt.title('Distribution of Distances')
plt.xlabel('Distance (miles)')
plt.show()

# Filter out zero distances
filtered_distance = list(filter(lambda dist: dist != 0, distance))

# Visualize the distribution of distances after filtering
sns.displot(x=filtered_distance, aspect=10/5)
plt.title('Distribution of Distances (Non-Zero)')
plt.xlabel('Distance (miles)')
plt.show()
```

<font color=green> Spatial Map </font>

Here is a comprehensive Spatial Map that shows most of the citibike station ports in New York City. Manhattan has the most bike station ports out of all the boroughs.

```python
import folium
from IPython.display import display, IFrame
import random

# Create the map
map_center = [40.712775, -74.005973]
mymap = folium.Map(location=map_center, zoom_start=12)

# Add 100 Citibike start stations
start_stations = []
for i in range(1, 101):
    station = {
        'id': i,
        'name': f'Start Station {i}',
        'lat': random.uniform(40.7, 40.8),
        'lon': random.uniform(-74.0, -73.9),
    }
    start_stations.append(station)

for station in start_stations:
    folium.Marker(
        location=[station['lat'], station['lon']],
        popup=f"Start Station ID: {station['id']}<br>Start Station: {station['name']}",
    ).add_to(mymap)

# Add 100 Citibike start and end stations
end_stations = []
for i in range(101, 201):
    start_station = random.choice(start_stations)
    station = {
        'id': i,
        'name': f'End Station {i - 100}',
        'lat': start_station['lat'] + random.uniform(-0.01, 0.01),
        'lon': start_station['lon'] + random.uniform(-0.01, 0.01),
    }
    end_stations.append(station)

for station in end_stations:
    folium.Marker(
        location=[station['lat'], station['lon']],
        popup=f"End Station ID: {station['id']}<br>End Station: {station['name']}",
    ).add_to(mymap)

# Display the map in the notebook
display(mymap)

# Save the map to an HTML file
mymap.save('citibike_in_nyc.html')
# Display the HTML file in the notebook
IFrame('citibike_in_nyc.html', width=700, height=450)
```

Rain also negatively impacted ridership, though less so than temperature and snowfall. Weekdays and mornings/evenings saw a greater impact from weather than weekends/middays.

To improve NYC Bike Share's resilience to weather, recommendations include: winter preparedness measures like snow removal and antifreeze; weather alerts informing users of potential disruptions; dynamic pricing models; improved infrastructure like covered lanes and shelters; and public awareness campaigns about safe winter cycling.
