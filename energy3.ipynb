# Methodolgy/Approach

**Data Source**

I used the World Energy Consumption data set found on kaggle.com. It is an extensive data set that is routinely updated from Data on Energy by *Our World in Data*. I used the csv file version and pandas dataframes to work with the data. I chose to take a close look at the global GDP leaders because I assumed they would have more robust options for types of energy available to consume.

Housekeeping

```python
import csv
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

importing data and creating dataFrame

```python
energyC_df = pd.read_csv('/content/owid-energy-data.csv')
```

filter columns

```python
energyC_df = energyC_df[['country','year','coal_production','electricity_generation','biofuel_electricity',
                       'coal_electricity','fossil_electricity','gas_electricity','hydro_electricity',
                       'nuclear_electricity','oil_electricity','renewables_electricity','oil_production',
                       'population','gdp','solar_electricity','wind_electricity','energy_per_gdp',
                       'energy_per_capita','fossil_share_elec','gas_share_elec','gas_production',
                       'low_carbon_share_elec']]
```

list of top 10 countries based on GDP

```python
top_gdp = ['United States','China','Japan','Germany','United Kingdom','India','France','Italy',
              'Canada','South Korea']
```

filter df to only show countries with highest gdp

```python
energyC_df=energyC_df.loc[energyC_df['country'].isin(top_gdp)]
```

filter df to show years 1990 onward

```python
energyC_df =energyC_df[energyC_df['year']>=1990]
```

```python
energyC_df
```

**Clean data**<br>
search for missing values and count


```python
nulls=energyC_df.isna().sum()
nulls
```

replace missing values

```python
nulls = energyC_df.isna().sum()
for i in nulls.index:
    if nulls[i]>0:
        energyC_df[i].ffill(inplace=True)
        energyC_df[i].bfill(inplace=True)
energyC_df.isna().sum()
```

```python
energyC_df
```

Groupby data into new dfs for plotting

```python
gdp_df =energyC_df.groupby(['year','country'])['gdp'].sum().unstack('country')
country_pop_df =energyC_df.groupby(['year','country'])['population'].sum().unstack('country')
pop_mil_df = round(country_pop_df/1000000)
energyC_df['gdp_per_capita'] = round(energyC_df['gdp']/energyC_df['population'])
perCap_gdp_df = energyC_df.groupby(['year','country'])['gdp_per_capita'].sum().unstack('country')
```

#Results

```python
sns.set_palette("tab10")
fig, ax = plt.subplots(3, 1, figsize=(24, 26))
ax[0].plot(gdp_df,label=gdp_df.columns)
ax[0].legend(loc="upper left")
ax[1].plot(pop_mil_df ,label=pop_mil_df .columns)
ax[1].legend(loc="upper left")
ax[2].plot(perCap_gdp_df,label=perCap_gdp_df.columns)
ax[2].legend(loc="upper left")
fig.text(0.1, 0.95, 'GDP and Population Per Country', fontsize=15, fontweight='bold')
fig.text(0.04, 0.5, '''mill''', fontsize=12, fontweight='light')
fig.text(0.5, 0.89, '''GDP''', fontsize=12, fontweight='light')
fig.text(0.5, 0.62,'''Population''', fontsize=12, fontweight='light')
fig.text(0.5, 0.35,'''GDP per capita''', fontsize=12, fontweight='light')
plt.show()
```

**Insights**

While looking at GDP and population for desired countries, we see that the United States has the highest GDP and the highest GDP per capita. Worth noting that China and India have the highest populations, but the lowest GDP per capita.

```python
energyC_df_US['electricity_generation'].plot(kind='line', figsize=(8, 4), title='electricity_generation')
plt.gca().spines[['top', 'right']].set_visible(False)
```

```python
def _plot_series(series, series_name, series_index=0):
  palette = list(sns.palettes.mpl_palette('Dark2'))
  xs = series['population']
  ys = series['electricity_generation']

  plt.plot(xs, ys, label=series_name, color=palette[series_index % len(palette)])

fig, ax = plt.subplots(figsize=(10, 5.2), layout='constrained')
df_sorted = energyC_df_US.sort_values('population', ascending=True)
_plot_series(df_sorted, '')
sns.despine(fig=fig, ax=ax)
plt.xlabel('population')
_ = plt.ylabel('electricity_generation')
```

```python
energyC_df_US_reset = energyC_df_US.reset_index()
ax = energyC_df_US_reset[['year', 'biofuel_electricity', 'coal_electricity',
                          'fossil_electricity', 'gas_electricity', 'hydro_electricity', 'nuclear_electricity', 'oil_electricity',
                          'renewables_electricity', 'solar_electricity', 'wind_electricity']].plot(kind='line', x='year', figsize=(16, 8), title='US Energy Consumption by Type')

plt.gca().spines[['top', 'right']].set_visible(False)

# Move legend outside the graph to the right (adjust the values as needed)
ax.legend(loc='upper left', bbox_to_anchor=(1, 1))

plt.show()
```

**Insight**<br>
In the US we can see that energy production has been climbing over the last 30 years. We can aslo see that as population rises in the US, so does energy consumption. While coal electricity consumption is down in the US, we can see that gas electricity consumption is trending upwards after 2010. Around 2015, there is a slight upward trend for solar, wind, and renewable energy.
